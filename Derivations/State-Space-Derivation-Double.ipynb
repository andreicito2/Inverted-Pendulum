The Lagrangian $(\mathcal{L})$ is the difference between the kinetic energy $( T )$ and potential energy $( V )$ of the system:
$$
\mathcal{L} = T - V
$$

**System Description:**
- $ x $: horizontal position of the cart
- $ \theta_1 $, $ \theta_2 $: angle of pendulums 1 and 2 from the vertical (upwards is $ \theta_i = 0 $)
- $ M $: mass of the cart
- $m_1$, $m_2$: Masses of pendulums 1 and 2
- $l_1$, $l_2$: Lengths to the center of mass of pendulums 1 and 2
- $I_1$, $I_2$: Moments of inertia of pendulums 1 and 2 about their centers of mass
- $ g $: acceleration due to gravity

**Kinetic Energy $( T )$:**
$$
T = \frac{1}{2} M \dot{x}^2 + \frac{1}{2} m_1 \left[ (\dot{x} + l_1 \dot{\theta_1} \cos\theta_1)^2 + (l_1 \dot{\theta_1} \sin\theta_1)^2 \right] + \frac{1}{2}I_1 \dot{\theta_1}^2 + \frac{1}{2} m_2 \left[ (\dot{x} + l_2 \dot{\theta_2} \cos\theta_2)^2 + (l_2 \dot{\theta_2} \sin\theta_2)^2 \right] + \frac{1}{2}I_2 \dot{\theta_2}^2
$$
Expanding and simplifying:
$$
T = \frac{1}{2} (M + m_1 + m_2) \dot{x}^2 + m_1 l_1 \dot{x} \dot{\theta_1} \cos\theta_1 + \frac{1}{2} (m_1 l_1^2 + I_1) \dot{\theta_1}^2 + m_2 l_2 \dot{x} \dot{\theta_2} \cos\theta_2 + \frac{1}{2} (m_2 l_2^2 + I_2) \dot{\theta_2}^2
$$

**Potential Energy $( V )$:**
$$
V = g (m_1 l_1 \cos\theta_1 + m_2 l_2 \cos\theta_2)
$$

**Lagrangian $(\mathcal{L})$:**
$$
\mathcal{L} = T - V
$$

$$
\mathcal{L} = \frac{1}{2} (M + m_1 + m_2) \dot{x}^2 + m_1 l_1 \dot{x} \dot{\theta_1} \cos\theta_1 + \frac{1}{2} (m_1 l_1^2 + I_1) \dot{\theta_1}^2 + m_2 l_2 \dot{x} \dot{\theta_2} \cos\theta_2 + \frac{1}{2} (m_2 l_2^2 + I_2) \dot{\theta_2}^2 - g (m_1 l_1 \cos\theta_1 + m_2 l_2 \cos\theta_2)
$$

**Euler-Lagrange Equations:**
The Euler-Lagrange equations for the system are given by:
$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{q_i}} \right) - \frac{\partial \mathcal{L}}{\partial q_i} = Q_i
$$
where $ q_i $ represents the generalized coordinates $ x $ and $ \theta $, and $ Q_i $ are the generalized forces.

**For x:**
$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} = u
$$
Calculating the derivatives:
$$
\frac{\partial \mathcal{L}}{\partial \dot{x}} = (M + m_1 + m_2) \dot{x} + m_1 l_1 \dot{\theta_1} \cos\theta_1 + m_2 l_2 \dot{\theta_2} \cos\theta_2
$$
$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) = (M + m_1 + m_2) \ddot{x} + m_1 l_1 \ddot{\theta_1} \cos\theta_1 - m_1 l_1 \dot{\theta_1}^2 \sin\theta_1 + m_2 l_2 \ddot{\theta_2} \cos\theta_2 - m_2 l_2 \dot{\theta_2}^2 \sin\theta_2
$$
$$
\frac{\partial \mathcal{L}}{\partial x} = 0
$$

Thus, the equation becomes:
$$
(M + m_1 + m_2) \ddot{x} + m_1 l_1 \ddot{\theta_1} \cos\theta_1 - m_1 l_1 \dot{\theta_1}^2 \sin\theta_1 + m_2 l_2 \ddot{\theta_2} \cos\theta_2 - m_2 l_2 \dot{\theta_2}^2 \sin\theta_2 = u
$$

**For $\theta_i$:**
$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{\theta}} \right) - \frac{\partial \mathcal{L}}{\partial \theta} = 0
$$
Calculating the derivatives:
$$
\frac{\partial \mathcal{L}}{\partial \dot{\theta}} = (m_i l_i^2 + I_i) \dot{\theta_i} + m_i l_i \dot{x_i} \cos\theta_i
$$
$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{\theta}} \right) = (m_i l_i^2 + I_i) \ddot{\theta} + m_i l_i \ddot{x} \cos\theta_i - m_i l_i \dot{x} \dot{\theta_i} \sin\theta_i
$$
$$
\frac{\partial \mathcal{L}}{\partial \theta} = -m_i l_i \sin\theta_i (\dot{x} \dot{\theta_i} - g)
$$

Thus, the equations become:
$$
(m_1 l_1^2 + I_1) \ddot{\theta_1} + m_1 l_1 \ddot{x} \cos\theta_1 - m_1 l_1 g \sin\theta_1 = 0
$$

And:
$$
(m_2 l_2^2 + I_2) \ddot{\theta_2} + m_2 l_2 \ddot{x} \cos\theta_2 - m_2 l_2 g \sin\theta_2 = 0
$$

**Linearize the Equations of Motion:**
To linearize the equations of motion, we assume small angles for the pendulum, which allows us to use the approximations $ \sin\theta \approx \theta $, $ \cos\theta \approx 1 $, and neglect higher-order terms.
The equations of motion become:
1. For x:
$$
(M + m_1 + m_2) \ddot{x} + m_1 l_1 \ddot{\theta_1} + m_2 l_2 \ddot{\theta_2} = u
$$
2. For $\theta_i$:
$$
(m_i l_i^2 + I_i) \ddot{\theta_i} + m_i l_i \ddot{x} - m_i l_i g \theta_i = 0
$$

**Add Damping Terms:**
To include damping, we add terms proportional to the velocities:
1. For x:
$$
(M + m_1 + m_2) \ddot{x} + c_x \dot{x} + m_1 l_1 \ddot{\theta_1} + m_2 l_2 \ddot{\theta_2} = u
$$
2. For $\theta_i$:
$$
(m_i l_i^2 + I_i) \ddot{\theta_i} + c_{\theta i} \dot{\theta} + m_i l_i \ddot{x} - m_i l_i g \theta_i = 0
$$

**Solve for Accelerations in Matrix Form:**
First we rearrange the equations to isolate the accelerations $ \ddot{x} $, $ \ddot{\theta_1} $, and $ \ddot{\theta_2} $:
1. For $ \ddot{x} $:
$$
(M + m_1 + m_2) \ddot{x} + m_1 l_1 \ddot{\theta_1} + m_2 l_2 \ddot{\theta_2} = u - c_x \dot{x}
$$

2. For $ \ddot{\theta_i} $:

$$
- m_i l_i \ddot{x} - (m_i l_i^2 + I_i) \ddot{\theta_i} = m_i l_i g \theta_i + c_{\theta i} \dot{\theta}
$$

To express the equations in matrix form, we can write:
$$\begin{bmatrix}
(M + m_1 + m_2) & m_1 l_1 & m_2 l_2 \\
- m_1 l_1 & - (m_1 l_1^2 + I_1) & 0 \\
- m_2 l_2 & 0 & - (m_2 l_2^2 + I_2)
\end{bmatrix}
\begin{bmatrix}
\ddot{x} \\
\ddot{\theta_1} \\
\ddot{\theta_2}
\end{bmatrix}
=
\begin{bmatrix}
u - c_x \dot{x} \\
m_1 l_1 g \theta_1 + c_{\theta 1} \dot{\theta_1} \\
m_2 l_2 g \theta_2 + c_{\theta 2} \dot{\theta_2}
\end{bmatrix}
$$

We can solve for the accelerations using the inverse of the coefficient matrix M:
$$
M = \begin{bmatrix}
(M + m_1 + m_2) & m_1 l_1 & m_2 l_2 \\
- m_1 l_1 & - (m_1 l_1^2 + I_1) & 0 \\
- m_2 l_2 & 0 & - (m_2 l_2^2 + I_2)
\end{bmatrix}
$$

Finally, we can express the accelerations $ \ddot{x} $ and $ \ddot{\theta} $ as:
$$
\begin{bmatrix}
\ddot{x} \\
\ddot{\theta_1} \\
\ddot{\theta_2}
\end{bmatrix}
=
M^{-1}
\begin{bmatrix}
u - c_x \dot{x} \\
m_1 l_1 g \theta_1 + c_{\theta 1} \dot{\theta_1} \\
m_2 l_2 g \theta_2 + c_{\theta 2} \dot{\theta_2}
\end{bmatrix}
$$

We can now evaluate the separate components of the accelerations:
1. For $ \ddot{x} $:
$$
\ddot{x} = M_{1,1}^{-1}(u - c_x \dot{x}) + M_{1,2}^{-1}(m_1 l_1 g \theta_1 + c_{\theta 1} \dot{\theta_1}) + M_{1,3}^{-1}(m_2 l_2 g \theta_2 + c_{\theta 2} \dot{\theta_2})
$$
$$
\ddot{x} = - M_{1,1}^{-1} c_x \dot{x} + M_{1,2}^{-1} m_1 l_1 g \theta_1 + M_{1,2}^{-1} c_{\theta 1} \dot{\theta_1} + M_{1,3}^{-1} m_2 l_2 g \theta_2 + M_{1,3}^{-1} c_{\theta 2} \dot{\theta_2} + M_{1,1}^{-1} u
$$


2. For $ \ddot{\theta_1} $:
$$
\ddot{\theta_1} = M_{2,1}^{-1}(u - c_x \dot{x}) + M_{2,2}^{-1}(m_1 l_1 g \theta_1 + c_{\theta 1} \dot{\theta_1}) + M_{2,3}^{-1}(m_2 l_2 g \theta_2 + c_{\theta 2} \dot{\theta_2})
$$
$$
\ddot{\theta_1} = - M_{2,1}^{-1} c_x \dot{x} + M_{2,2}^{-1} m_1 l_1 g \theta_1 + M_{2,2}^{-1} c_{\theta 1} \dot{\theta_1} + M_{2,3}^{-1} m_2 l_2 g \theta_2 + M_{2,3}^{-1} c_{\theta 2} \dot{\theta_2} + M_{2,1}^{-1} u
$$

3. For $ \ddot{\theta_2} $:
$$
\ddot{\theta_2} = M_{3,1}^{-1}(u - c_x \dot{x}) + M_{3,2}^{-1}(m_1 l_1 g \theta_1 + c_{\theta 1} \dot{\theta_1}) + M_{3,3}^{-1}(m_2 l_2 g \theta_2 + c_{\theta 2} \dot{\theta_2})
$$
$$
\ddot{\theta_2} = - M_{3,1}^{-1} c_x \dot{x} + M_{3,2}^{-1} m_1 l_1 g \theta_1 + M_{3,2}^{-1} c_{\theta 1} \dot{\theta_1} + M_{3,3}^{-1} m_2 l_2 g \theta_2 + M_{3,3}^{-1} c_{\theta 2} \dot{\theta_2} + M_{3,1}^{-1} u
$$

Finally, we can create the state-space matrices A and B for the system:
$$
A = \begin{bmatrix}
0 & 1 & 0 & 0 & 0 & 0 \\
0 & - M_{1,1}^{-1} c_x & M_{1,2}^{-1} m_1 l_1 g & M_{1,2}^{-1} c_{\theta 1} & M_{2,3}^{-1} m_2 l_2 g & M_{2,3}^{-1} c_{\theta 2} \\
0 & 0 & 0 & 1 & 0 & 0 \\
0 & - M_{2,1}^{-1} c_x & M_{2,2}^{-1} m_1 l_1 g & M_{2,2}^{-1} c_{\theta 1} & M_{2,3}^{-1} m_2 l_2 g & M_{2,3}^{-1} c_{\theta 2} \\
0 & 0 & 0 & 0 & 0 & 1 \\
0 & - M_{3,1}^{-1} c_x & M_{3,2}^{-1} m_1 l_1 g & M_{3,2}^{-1} c_{\theta 1} & M_{3,3}^{-1} m_2 l_2 g & M_{3,3}^{-1} c_{\theta 2} \\
\end{bmatrix}
$$
$$
B = \begin{bmatrix}
0 \\
M_{1,1}^{-1} \\ 
0 \\ 
M_{2,1}^{-1} \\ 
0 \\
M_{3,1}^{-1} \\ 
\end{bmatrix}
$$
